```python
# 18/5/24
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos
xn= np.array([0.00, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 1.00])
yn= np.array([1.00, 1.64, 1.51, 2.03, 2.75, 3.59, 4.87, 5.23, 5.44, 6.37])
```

```python
plt.scatter(xn, yn, marker='*') 
plt.title(r'Velocidad vs tiempo', fontsize=20) 
plt.xlabel(r'$t$ [s]', fontsize=16)
plt.ylabel(r'$v$ [m/s]', fontsize=16)
```

```python
# Se obtiene el valor de n (numero de datos)
n = len(xn)
# Las sumatorias necesarias 
Sum_x = np.sum(xn)
Sum_y = np.sum(yn)
Sum_xx = np.sum(xn**2)
Sum_xy = np.sum(xn*yn)
Delta = n*np.sum(xn**2) - np.sum(xn)**2
print(n,',', Sum_x, ',',Sum_y,',', Sum_xx,',', Sum_xy)
```

```python
# Se escriben las ecuaciones para b y m 
m_mc = (n * Sum_xy - Sum_x * Sum_y) / Delta
b_mc = Sum_y /n - m_mc * Sum_x/n
print('m=',m_mc, ',', 'b=',b_mc)
```

```python
# La gráfica con los datos y la recta que mejor se ajusta 
y_pred= m_mc*xn + b_mc
# 
plt.figure()
plt.scatter(xn, yn, color='b',marker='+', label='datos medidos')
plt.plot(xn, y_pred, 'r--',label='recta por mínimos cuadrados')
plt.grid(linestyle='dotted')
plt.legend(loc='best')
plt.title(r'Velocidad vs tiempo', fontsize=18)
plt.xlabel(r'$t$ [s]', fontsize=16)
plt.ylabel(r'$v$ [m/s]', fontsize=16)
plt.text(0.6, 1.0, '$y=(5.88) x + 0.736$', fontsize=12)
plt.show()
```

```python
# Escribimos las sumatorias para calcular Sy
error = yn - y_pred
SSE = np.sum(error ** 2)
Sy = np.sqrt(SSE / (n - 2))
print("Error estándar de la estimación (RMSE):", Sy)
```

```python
Delta_m = np.sqrt(n/(n*np.sum(xn ** 2) - np.sum(xn)**2))*Sy
Delta_b = np.sqrt(np.sum(xn**2)/(n*np.sum(xn**2)-np.sum(xn)**2))*Sy
print(f'm = {np.round(m_mc, 1)} \u00B1 {np.round(Delta_m, 1)}')
print(f'b = {np.round(b_mc, 1)} \u00B1 {np.round(Delta_b, 1)}')
```

```python
# Incertidumbre (en este caso, usando un valor fijo para la simplicidad)
incertidumbre_y = 0.01  # Se ajusta este valor según tus necesidades

# Coeficientes de la recta de mejor ajuste
m_mc = (n * Sum_xy - Sum_x * Sum_y) / Delta
b_mc = Sum_y /n - m_mc * Sum_x/n

# Gráficamos los datos y la recta ajustada
x = xn
y = m_mc*x + b_mc

plt.figure()
plt.errorbar(xn, yn, yerr=incertidumbre_y, fmt='o', label='datos medidos')
plt.plot(x, y, 'r--', label='recta por mínimos cuadrados')
plt.grid(linestyle='dotted')
plt.legend(loc='best')
plt.xlabel('t [s]')
plt.ylabel('v [m/s]')
plt.text(0.6, 2.5, r'$y = ({:.1f})x + {:.2f}$'.format(m_mc, b_mc), fontsize=12)
plt.tight_layout()
plt.show()

# El cálculo de los errores 
y_pred = y
error = yn - y_pred
SSE = np.sum(error ** 2)
Sy = np.sqrt(SSE / (n - 2))

Delta_m = np.sqrt(n/(n*np.sum(xn ** 2) - np.sum(xn)**2))*Sy
Delta_b = np.sqrt(np.sum(xn**2)/(n*np.sum(xn**2)-np.sum(xn)**2))*Sy

print(f'm = {np.round(m_mc, 3)} \u00B1 {np.round(Delta_m, 3)}')
print(f'b = {np.round(b_mc, 3)} \u00B1 {np.round(Delta_b, 3)}')
```
