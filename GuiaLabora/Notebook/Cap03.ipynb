```python
# 1217/5/24
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.array ([1.780 , 1.780 , 1.780 ,1.790 , 1.790])
print('Num de datos:', len(d))
print('Promedio:', np.mean(d), 'cm')
print('Suma:', np.sum(d))
```

```python
# Valores máximo y mínimos del conjunto de datos
print('valor máximo:', np.amax(d))
print('valor mínimo:', np.amin(d))
print('dispersión:', (np.amax(d) - np.amin(d)).round(2))
print('dispersión porcentual:', ((np.amax(d)-np.amin(d))/np.mean(d)*100).round(1),'%')
```

```python
Delta_d=(np.sum(np.abs(np.mean(d)-d))/len(d)).round(3) 
Delta_d
```

```python
err_d=(Delta_d/np.mean(d)) # error relativo 
err_dp=(err_d*100) # error porcentual
print('error relativo:', err_d.round(4)) 
print('error porcentual:', err_dp.round(2),'%')
```
