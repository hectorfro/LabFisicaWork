```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos experimentales (Ejemplo)
tiempo = np.array([0, 1, 2, 3, 4, 5])  # Tiempo en segundos
voltaje_descarga = np.array([10, 7.36, 5.42, 4.00, 2.94, 2.17])  # Voltaje en la descarga

# Transformación logarítmica de los datos de voltaje de descarga
ln_voltaje_descarga = np.log(voltaje_descarga)

# Realizar regresión lineal sobre los datos transformados
slope, intercept, r_value, p_value, std_err = linregress(tiempo, ln_voltaje_descarga)

# Graficar los datos transformados y la recta de regresión
plt.scatter(tiempo, ln_voltaje_descarga, label='Datos experimentales')
plt.plot(tiempo, slope*tiempo + intercept, label=f'Ajuste: Pendiente={slope:.2f}', color='red')

# Mostrar resultados
plt.xlabel('Tiempo (s)')
plt.ylabel('ln(Voltaje) (ln(V))')
plt.title('Descarga del Capacitor - Transformación Logarítmica')
plt.legend()
plt.grid(True)
plt.show()

# Resultados numéricos
print(f'Pendiente calculada: {slope:.2f}')
print(f'Intercepto: {intercept:.2f}')
print(f'Valor p: {p_value:.5f}')
print(f'Constante de tiempo RC (calculada): {-1/slope:.2f} s')
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Datos experimentales (Ejemplo)
tiempo = np.array([0, 1, 2, 3, 4, 5])  # Tiempo en segundos
voltaje_descarga = np.array([10, 7.36, 5.42, 4.00, 2.94, 2.17])  # Voltaje en la descarga

# Transformación logarítmica de los datos de voltaje de descarga
ln_voltaje_descarga = np.log(voltaje_descarga)

# Agregar una constante para el término independiente (intercepto)
tiempo = sm.add_constant(tiempo)

# Crear el modelo de regresión lineal
modelo = sm.OLS(ln_voltaje_descarga, tiempo)
resultados = modelo.fit()

# Imprimir el resumen estadístico
print(resultados.summary())

# Graficar los datos transformados y la recta de regresión
plt.scatter(tiempo[:, 1], ln_voltaje_descarga, label='Datos experimentales')
plt.plot(tiempo[:, 1], resultados.predict(tiempo), label='Ajuste de regresión', color='red')

# Mostrar resultados
plt.xlabel('Tiempo (s)')
plt.ylabel('ln(Voltaje) (ln(V))')
plt.title('Descarga del Capacitor - Transformación Logarítmica')
plt.legend()
plt.grid(True)
plt.show()
```
