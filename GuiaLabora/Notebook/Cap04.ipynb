```python
# 21/5/24
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = 5*10**(1/3*x)
plt.plot(x, y, 'o')
plt.show()
```

```python
plt.plot(x, y, '-o')
plt.yscale("log")
plt.show()
```

```python
x = np.arange(20)
y = 8*x**(3)
plt.plot(x, y, '*')
plt.show()
```

```python
plt.plot(x, y, '-*')
plt.xscale("log")
plt.yscale("log")
plt.show()
```
